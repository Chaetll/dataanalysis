```python
import pandas as pd
```

```python
file_name = "https://people.sc.fsu.edu/~jburkardt/data/csv/homes.csv"
df = pd.read_csv(file_name)
```

```python
df.columns = ['Sell', 'List', 'Living', 'Rooms', 'Beds', 'Baths',
       'Age', 'Acres', 'Taxes']
```

```python
df.head()
```

```python
df.dtypes()
df.count()
df = df.drop_duplicates()
df.count()
```

```python
rooms = df.groupby(['Rooms']).mean()
```

```python
taxes = df[['Beds','Taxes']].groupby('Beds').mean().reset_index()
```

```python
taxes.head()
```

```python
df['large_beds'] = df['Beds'] >= 4
```

```python
df.head()
```

```python
lb = df[df['large_beds']= True]
```

```python
lb = df[df['large_beds'] == True].head()
```

```python
sb = df[df['large_beds'] == False].head()
```

```python
lb.head()
```

```python
sb.head()
```

```python
rooms.append(taxes)
```

```python
pd.concat([sb,lb])
```

```python
pd.concat([sb,taxes])
```

```python
#MERGE FUNCTIONS
```

```python
#Inner Join
pd.merge(df,taxes, on="Beds").head()
```

```python
#Left Join (if NULL)
pd.merge(df,taxes[taxes["Beds"] < 4], on="Beds", how='left').head()
```

```python
#Right Join
pd.merge(df,taxes[taxes["Beds"] < 4], on="Beds", how='right').head()
```

```python
#Full Outer Join
pd.merge(lb,taxes[taxes["Beds"] <= 4], on="Beds", how='outer').head()
```

```python
#Left Join
pd.merge(df,taxes, on="Beds", how='left').head()
```

```python
rbs = df.groupby(['Beds','Rooms'])['Taxes'].mean().reset_index()
```

```python
rbs = rbs.rename(columns={"Taxes":"avg_taxes"})
```

```python
pd.merge(df,rbs, on=["Beds","Rooms"], how='left').head()
```

```python
df.join(taxes[taxes["Beds"] < 4], on=["Beds"], how='left',lsuffix='_l')
```

```python
taxes[taxes["Beds"] < 4].join(df, on=["Beds"], how='right',lsuffix='_l')
```

```python
taxes[taxes["Beds"] < 4].join(df, on=["Beds"], how='inner',lsuffix='_l')
```
