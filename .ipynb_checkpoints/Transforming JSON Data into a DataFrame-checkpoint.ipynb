```python
#Requests used to GET data from an external API with JSON data
import requests
#Pandas for Data Analysis
import pandas as pd
#Pandas json_normalize function for reading transforming JSON data to Pandas DataFrame
from pandas.io.json import json_normalize
#JSON for loading data into a DataFrame
import json
```

```python
url = "https://gbfs.citibikenyc.com/gbfs/en/station_information.json"
r = requests.get(url)
```
