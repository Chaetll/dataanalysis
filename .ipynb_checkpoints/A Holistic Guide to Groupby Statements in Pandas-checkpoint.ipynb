```python
import pandas as pd
file_name = "https://people.sc.fsu.edu/~jburkardt/data/csv/homes.csv"
df = pd.read_csv(file_name)
```

```python
df.head()
```

```python
#Clean up the column header names - they include " and spaces as-is
df.columns = [col.strip(' ').strip('"') for col in df.columns]
```

```python
df.head()
```

```python
df.groupby('Beds')['Acres'].mean()
```

```python
#We can see the output here is a Pandas Series
type(df.groupby('Beds')['Acres'].mean())
```

```python
#Sort Argument
df.groupby(['Beds','Baths'],sort=0).mean()
```

```python
df.groupby(['Beds','Baths'],as_index=False).mean()
```

```python
df.groupby('Beds',as_index=False)['Acres'].mean()
```

```python
df.groupby('Beds')['Acres'].describe()
```

```python
#using the agg() function for multiple statistics
df.groupby(['Beds', 'Baths'])['Acres'].agg(['mean', 'count'])
```
